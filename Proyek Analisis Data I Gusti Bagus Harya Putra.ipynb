### Explore order_items_df, order_payments_df dan seller_df

```python
# Langkah 1: Gabungkan order_items_df dengan order_payments_df berdasarkan order_id
merged_orders = order_items_df.merge(
    order_payments_df,
    on='order_id',
    how='left'
)

# Langkah 2: Tambahkan kolom baru price_difference
merged_orders['price_difference'] = merged_orders['total_order_price'] - merged_orders['payment_value']

# Langkah 3: Gabungkan hasil dengan sellers_df berdasarkan seller_id
combined1_df = merged_orders.merge(
    sellers_df,
    on='seller_id',
    how='left'
)
combined1_df.head()
```

**Insight:**
- Dalam data ini, pesanan terindikasi ada perbedaan harga, terlihat di baris empat. Ini bisa menyebabkan ketidakpuasan pelanggan karena pembayaran tidak tepat

### Explore order_reviews_df

```python
# 1. Kelompokkan review_score
score_bins = [0, 2, 3, 5]
score_labels = ["Not Satisfied", "Neutral", "Satisfied"]
order_reviews_df['review_category'] = pd.cut(order_reviews_df['review_score'], bins=score_bins, labels=score_labels)

# 2. Tambahkan kolom review_comment_value
order_reviews_df['review_comment_value'] = order_reviews_df.apply(
    lambda row: 'Yes' if pd.notnull(row['review_comment_title']) or pd.notnull(row['review_comment_message']) else 'No',
    axis=1
)
order_reviews_df
```

**Insight:**
- Hasil ini menampilkan ada yang menulis review dan tidak beserta tingkat kepuasan pelanggan
- Hasil ini bisa divisualisasikan di tahap visualisasi data

### Explore orders_df

```python
# 1. Hitung selisih waktu antara order_purchase_timestamp dan order_approved_at
orders_df['approval_time_diff'] = (orders_df['order_approved_at'] - orders_df['order_purchase_timestamp']).dt.total_seconds() / 3600  # dalam jam

# 2. Hitung selisih waktu antara order_delivered_carrier_date dan order_delivered_customer_date
orders_df['delivery_time_diff'] = (orders_df['order_delivered_customer_date'] - orders_df['order_delivered_carrier_date']).dt.total_seconds() / 3600  # dalam jam

# 3. Bandingkan order_delivered_customer_date dengan order_estimated_delivery_date
orders_df['on_time_delivery'] = orders_df.apply(
    lambda row: 'Yes' if (row['order_estimated_delivery_date'] - row['order_delivered_customer_date']).days >= 0 else 'No',
    axis=1
)
orders_df
```

```python
print("Approval Time Difference:")
print("Mean:", orders_df['approval_time_diff'].mean())
print("Minimum:", orders_df['approval_time_diff'].min())
print("Maximum:", orders_df['approval_time_diff'].max())

print("\nDelivery Time Difference:")
print("Mean:", orders_df['delivery_time_diff'].mean())
print("Minimum:", orders_df['delivery_time_diff'].min())
print("Maximum:", orders_df['delivery_time_diff'].max())

# Calculate frequency of 'Yes' and 'No' in on_time_delivery
on_time_delivery_counts = orders_df['on_time_delivery'].value_counts()

print("\nOn-Time Delivery Frequency:")
print("Yes:", on_time_delivery_counts.get('Yes', 0))
print("No:", on_time_delivery_counts.get('No', 0))
```

**Insight:**
- Hasil ini menampilkan selisih waktu pembelian disetujui tidak terlalu tinggi karena rata-rata hanya di angka 10 jam. Menandakan hal ini bukan faktor utama dalam ketidakpuasan customer
- Hasil selisih waktu pengiriman menunjukkan rata-rata yang tinggi, menunjukkan bahwa paket dikirimkan sangat lama. Hal ini bisa menjadi salah satu faktor utama dalam ketidakpuasan pasien
- Hasil pengiriman yang sampai lebih awal bahkan tepat waktu tergolong banyak, namun banyak juga yang terlambat. Hal ini bisa menjadi salah satu faktor utama ketidakpuasan pasien
- Hasil ini bisa divisualisasikan di tahap visualisasi data nantinya

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana performa penjualan bisnis e-commerce dalam beberapa bulan terakhir?

1. Membuat dataframe bulanan

```python
# Resample data untuk mendapatkan jumlah pesanan unik dan total pendapatan per bulan
monthly_orders_df = alldata_df.resample(
    rule='M',
    on='order_purchase_timestamp'
).agg({
    "order_id": "nunique",  # Jumlah pesanan unik
    "total_order_price": "sum"  # Total pendapatan
})

# Ubah index menjadi format Tahun-Bulan
monthly_orders_df.index = monthly_orders_df.index.strftime('%Y-%B')

# Reset index agar lebih mudah diakses
monthly_orders_df = monthly_orders_df.reset_index()

# Ganti nama kolom agar lebih deskriptif
monthly_orders_df.rename(columns={
    "order_purchase_timestamp": "order_date",
    "order_id": "order_count",
    "total_order_price": "revenue"
}, inplace=True)

monthly_orders_df
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_orders_df["order_date"],
    monthly_orders_df["order_count"],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)
plt.title("Number of Orders per Month", loc="center", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Number of Orders", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(visible=True, linestyle="--", alpha=0.6)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(
    monthly_orders_df["order_date"],
    monthly_orders_df["revenue"],
    marker='o',
    linewidth=2,
    color="#72BCD4"
)
plt.title("Total Revenue per Month", loc="center", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Total Revenue (Currency)", fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.yticks(fontsize=10)
plt.grid(visible=True, linestyle="--", alpha=0.6)
plt.show()
```

**Insight:**
- Kedua tabel ini menampilkan banyaknya pesanan dan banyaknya pendapatan yang diperoleh dari bulan september tahun 2016 dan bulan oktober tahun 2018
- Grafik naik turun menunjukkan tren penjualan dan pendapatan bisnis e-commerce

### Pertanyaan 2: Bagaimana demografi pelanggan yang ada pada bisnis e-commerce ini?

#### Berdasarkan city (kota)

```python
# Kelompokkan berdasarkan city dan ambil 4 kota terbanyak
bycity_df = alldata_df.groupby(by="customer_city").customer_id.nunique().reset_index()
bycity_df.rename(columns={"customer_id": "customer_count"}, inplace=True)

# Ambil 4 kota terbanyak
top_4_cities_df = bycity_df.sort_values(by="customer_count", ascending=False).head(4)
top_4_cities_df
```

#### Berdasarkan states (provinsi)

```python
# Kelompokkan berdasarkan state dan ambil 4 state terbanyak
bystate_df = alldata_df.groupby(by="customer_state").customer_id.nunique().reset_index()
bystate_df.rename(columns={"customer_id": "customer_count"}, inplace=True)

# Ambil 4 state terbanyak
top_4_states_df = bystate_df.sort_values(by="customer_count", ascending=False).head(4)
top_4_states_df
```

```python
# Visualisasi bar plot untuk 4 provinsi terbanyak
plt.figure(figsize=(10, 6))
sns.barplot(
    y="customer_count",
    x="customer_state",
    data=top_4_states_df,
    palette="Blues_d"  # Pilihan palette warna
)
plt.title("Top 4 States with Most Customers", loc="center", fontsize=16)
plt.ylabel("Customer Count", fontsize=12)
plt.xlabel("State", fontsize=12)
plt.tick_params(axis='x', labelsize=10, rotation=45)  # Rotate labels on x-axis
plt.show()
```

**Insight:**
- Kedua chart ini menggambarkan demografi pelanggan dengan kota dan provinsi tertinggi

### Pertanyaan 3: Mengapa beberapa customer tidak menulis review? Apa faktornya?

#### Visualisasi bar chart untuk melihat distribusi value dari review comment

```python
# Visualisasi bar chart untuk distribusi review_comment_value (yes/no)
plt.figure(figsize=(8, 6))
sns.countplot(data=alldata_df, x="review_comment_value", palette="viridis")
plt.title("Distribution of Review Comment Value (Yes/No)", loc="center", fontsize=16)
plt.ylabel("Frequency", fontsize=12)
plt.xlabel("Review Comment Value (Yes/No)", fontsize=12)
plt.show()
```

#### Visualisasi line plot untuk selisih waktu pembelian diterima

```python
# Mengambil bulan dan tahun dari order_purchase_timestamp
alldata_df['month_year'] = alldata_df['order_purchase_timestamp'].dt.to_period('M')

# Menghitung rata-rata approval_time_diff per bulan
monthly_approval_time_diff = alldata_df.groupby('month_year')['approval_time_diff'].mean().reset_index()

# Visualisasi line plot untuk rata-rata approval_time_diff per bulan
plt.figure(figsize=(10, 6))
plt.plot(monthly_approval_time_diff['month_year'].astype(str), monthly_approval_time_diff['approval_time_diff'], marker='o', linewidth=2, color="turquoise")
plt.title("Average Approval Time Difference per Month", loc="center", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Average Approval Time Difference (Hours)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

#### Visualisasi line plot untuk selisih waktu pengiriman paket

```python
# Menghitung rata-rata delivery_time_diff per bulan
monthly_delivery_time_diff = alldata_df.groupby('month_year')['delivery_time_diff'].mean().reset_index()

# Visualisasi line plot untuk rata-rata delivery_time_diff per bulan
plt.figure(figsize=(10, 6))
plt.plot(monthly_delivery_time_diff['month_year'].astype(str), monthly_delivery_time_diff['delivery_time_diff'], marker='o', linewidth=2, color='r')
plt.title("Average Delivery Time Difference per Month", loc="center", fontsize=16)
plt.xlabel("Month", fontsize=12)
plt.ylabel("Average Delivery Time Difference (Days)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

#### Visualisasi bar chart untuk melihat distribusi pengiriman paket yang aman (sampai lebih cepat/on time) dan paket yang gagal (terlambat)
